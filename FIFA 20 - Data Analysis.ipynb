# 11. Printing who has 5 stars at Skill Moves.

```python
def bestskills(skill_moves):  #skill_moves is an int value and its max value is 5.
    if skill_moves > 4:
        return True
    return False
fifa[fifa["skill_moves"].apply(bestskills)]
```

# 12. Printing players whose contract ends in 2021.

```python
def contract_ends(contract_valid_until):
    if contract_valid_until == 2021:
        return True
    return False
fifa[fifa["contract_valid_until"].apply(contract_ends)]
```

# 14. Which position is most valuable?

```python
players=fifa.groupby("team_position")["value_eur"].mean()/1e6
players=players.sort_values()
players.plot(kind="barh",figsize=(12,8))
plt.xlabel("Average value (M euro)")
```

# 15. Age-based value chart

```python
ages=fifa.groupby("age")["value_eur"].mean()/1e6
ages.plot(grid=True,figsize=(12,8))
plt.ylabel("Average value (M euro)")
plt.xlabel("Age")
```

# 16. Finding most valuable National Teams.

```python
national_teams=fifa.groupby("nationality")["value_eur"].nlargest(25).reset_index(level=1, drop=True)
national_teams=national_teams.groupby("nationality").mean()
bests=(national_teams.sort_values()/1e6).tail(10)
bests.plot(kind="barh",figsize=(12,8))
plt.xlabel("Average value of best 25 players (M euro)")
```

# 19. Graph of players' age.

```python
sns.set(style = "dark", palette = "bright", color_codes = True)
ages = fifa.age
plt.figure(figsize = (15,8))
ax = sns.distplot(ages, bins = 58, kde = False, color = "green")
ax.set_xlabel(xlabel = "Player\'s age", fontsize = 15)
ax.set_ylabel(ylabel = "Number of players", fontsize = 15)
ax.set_title(label = "Graph of players\' age", fontsize = 20)
plt.show()
```
